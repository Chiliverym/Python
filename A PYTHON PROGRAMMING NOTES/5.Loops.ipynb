### Write a program to print Hello World 5 times using while loop.

```python
b=0
while b<5:
    print("Hello World")
    b=b+1
```

- This is because x start with 0, as long as the condition x<5 is True, the while loop keeps looping & inside the while loop x variable is increased by 1 , at the end of the 1st loop the execution jumps back to the start of the while loop and rechecks the condition and if the condition is still True it re-enters the block of code again .
- So this is why, the block runs 5 times but, at the 5th time the x is set to the value 5 & the while loop condition x<5 becomes False .So the execution stops after the condition becomes False and the program gets end.
- When the execution is runs through a loop its called a iteration .So we say that the while loop iterates 5 times, on each iteration the x variable increased by 1.

## If vs while

```python
x=0
if x < 5:
    print('Hello world.')
    x = x + 1
```

```python
x = 0
while x < 5:
    print('Hello world.')
    x = x + 1
```

- Similar to the if statement , but the difference is " How they behave?" at the end of an 'if' block the program execution continues on with the rest of the program .
- But at the end of the while block the execution jumps back to the start of the while loop & re-check the condition. So, if the condition is still True then the execution re-enters the while block , it keeps doing this & looping around through this code until the condition gets False, & then continues on the rest of the program.

### **Break Statement:**
The break statement causes the execution to immediately jumps out of the loop, it  doesn't even checks the condition again , it just immediately jumps out to execute the next line after of the loop.


```python
name= ''
while True:
    print("Please type ur name.")
    name=input()
    if name == 'jay':
        break
print("Thank you")
```

The break statement causes the executions to jumps out of the loop and then continue on with the rest of the program & prints Thank you

## While Continue

### Continue statement:
Like break statement , the continue statement are also used inside the loop & when the program execution reaches to the condition statement the execution immediately jumps back to the start of the loop and re-evaluate the loop condition.

```python
x=0
while(x<=10):
    x=x+1
    if(x==5):
        continue
    print(x)
```

- x is 5 , is missing that's because the when x is set 5, The if condition gets True and the continue statement is executed .
- So when continue is executed the execution immediately jumps back to the start of the while loop , so in this case that print function never happens. 
- That's why, you see when the execution is runs through a loop, we called that is a iteration.

```python
#Program for prime number

n = int(input('Enter a number-: '))
c=0

for i in range(2,n):
    if(n%i==0):
        c=c+1
        break
if(c==0):
    print(f"{n} is a prime number")
else:
    print(f"{n} is not a prime number")
```

```python
#Program for perfect number

n = int(input('Enter a number-: '))
tot=0
for i in range(1,n):
    if(n%i==0):
        tot=tot+i
        
if(tot==n):
    print(f"{n} is a perfect number")
else:
    print(f"{n} is not a perfect number")
```

```python
#Program for armstrong

n= int(input('Enter a Number-: '))
n1=n
n2=n
tot=0
d = 0      #d= len(str(n))  

while(n!=0):
    n=n//10
    d=d+1

while(n1!=0):
    rem = n1%10
    tot = tot + (rem**d)
    n1=n1//10

if(n2==tot):
    print(f"{n2} is Armstrong Number")
else:
    print(f"{n2} is not Armstrong Number")
```

```python
#Program for strong number

n=int(input("Enter a number-: "))
n1=n

tot = 0
while(n!=0):
    rem = n % 10

    fact = 1
    for i in range(1,rem+1):
        fact = fact * i
    
    tot = tot + fact
    n=n//10


if(n1==tot):
    print(f"{n1} is strong no")
else:
    print(f"{n1} is not strong no")
```

```python
#Program for neon number

n=int(input("Enter a number-: "))

sq=n*n
tot=0

while(sq!=0):
    rem=sq%10
    tot = tot + rem
    sq = sq // 10

if(n==tot):
    print(f"{n} is Neon number")
else:
    print(f"{n} is not Neon number")
```

# Fibonacci Series

```python
previous=0
current=1

print("The Fibonacci Series is : ")

while current < 501:
    print(current,end=" ")
    previous,current=current,previous+current
```
